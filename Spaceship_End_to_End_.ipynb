```python
# in Cabin we have some missing value and we have to deal with it and also split this column so weare going to create a spliting function

def split_cabin(x):
  if len(str(x).split('/')) < 3:
    return ['Missing', 'Missing']
  else:
     return str(x).split('/')
```

```python
# lets create a function that will be used for preprocessing 

def preprocessing(df):
  # Fill missing value in homeplant with missing
  df['HomePlanet'].fillna('missing')
  df['CryoSleep'].fillna('Missing', inplace = True)
  # cabin preprocessing - extract deck and side 
  df['TempCabin'] = df['Cabin'].apply(lambda x: split_cabin(x))
  df['deck'] = df['TempCabin'].apply(lambda x: x[0])
  df['side'] = df['TempCabin'].apply(lambda x: x[2] if len(x) >= 3 else 'Missing')
  df.drop(['TempCabin', 'Cabin'], axis = 1, inplace=True)
  # destination
  df['Destination'].fillna('missing', inplace = True)
  # age 
  df['Age'].fillna(df['Age'].mean(), inplace = True)
  # VIP - filling the row rows
  df['VIP'].fillna('missing', inplace = True)

  # menetary spending columns
  df['RoomService'].fillna(0, inplace = True)
  df['FoodCourt'].fillna(0, inplace = True)
  df['ShoppingMall'].fillna(0, inplace = True)
  df['Spa'].fillna(0, inplace = True)
  df['VRDeck'].fillna(0, inplace = True)
  # droping the name column
  df.drop('Name',axis=1, inplace=True)
  # droping the remaining rows
  # df.dropna(inplace = True)
```

# Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
X = data.drop(['Transported','PassengerId'] ,axis = 1)
X = pd.get_dummies(X)
y = data['Transported']
```

```python
X.head(2)
```

#SetUp ML pipeline

```python
from sklearn.ensemble import GradientBoostingClassifier , RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y , test_size= 0.3, random_state=1234) 
```

```python
pipelines = {
    'rf': make_pipeline(StandardScaler(), RandomForestClassifier(random_state=1234)),
    'gb': make_pipeline(StandardScaler(), GradientBoostingClassifier(random_state=1234))
}

# Define hyperparameters for each algorithm
grid = {
    'rf': {
        'randomforestclassifier__n_estimators': [100, 200, 300]
    },
    'gb': {
        'gradientboostingclassifier__n_estimators': [100, 200, 300]
    }
}
```

```python
fit_model = {}
for algo, pipeline in pipelines.items():
    if algo == 'rf':
        param_grid = {'randomforestclassifier__n_estimators': [100, 200, 300]}
    elif algo == 'gb':
        param_grid = {'gradientboostingclassifier__n_estimators': [100, 200, 300]}
    model = GridSearchCV(pipeline, param_grid, cv=5, n_jobs=-1, scoring='accuracy')
    model.fit(X_train[:1000], y_train[:1000])  # Use a smaller training set
    fit_model[algo] = model.best_estimator_


# fit_model = {}
# for algo, pipeline in pipelines.items():
#     model = GridSearchCV(pipeline, grid[algo], cv=5, n_jobs=-1, scoring='accuracy')
#     model.fit(X_train, y_train)
#     fit_model[algo] = model
```

#Evalutating the model

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
for algo, model in fit_model.items():
    y_pred = model.predict(X_test)
    accuracy = model.score(X_test, y_test)
    print(f"{algo}: {accuracy}")
```
